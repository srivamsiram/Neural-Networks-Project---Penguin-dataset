```python
df.dropna(inplace=True)
df.head()
```

```python
df1 = df.copy()
df1.head()
```

```python
Y = df1.species
Y.head()
```

```python
Y.unique()
```

```python
Y = Y.map({'Adelie': 0, 'Chinstrap':1, 'Gentoo':2})
Y.head()
```

```python
#### Drop the species column from df1

df1.drop('species', inplace=True, axis=1)
```

```python
df1.head()
```

```python
pd.get_dummies(df1['sex']).head()
```

```python
se=pd.get_dummies(df1['sex'], drop_first=True)
se.head(3)
```

```python
df1.island.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1['island']= le.fit_transform(df1['island'])

df1['island']
```

```python
df1.head()
```

```python
df2 = pd.concat([df1, se], axis=1)
```

```python
df2.head()
```

```python
df2.drop('sex', axis=1, inplace=True);
```

```python
df2.dtypes
```

```python
Y.dtypes
```

```python
X = df2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=40)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)

model1 = rfc.fit(X_train, y_train)
prediction1 = model1.predict(X_test)
print("Accuracy of Randomforest Classifier is:",accuracy_score(y_test,prediction1))
print("Precision Score of Randomforest Classifier is:",precision_score(y_test, prediction1,average='weighted'))
print("F1 Score is : ",f1_score(y_test, prediction1, average='weighted'))
```

```python
print(confusion_matrix(y_test, prediction1))
```

```python
print(classification_report(y_test, prediction1))
```

```python
df_val = pd.DataFrame({'Actual': y_test, 'Predicted_Score': prediction1})
df_val
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
KNN.fit(X_train, y_train)
prediction2= KNN.predict(X_test)
print("Accuracy of KNeighborsClassifier is:",accuracy_score(y_test, prediction2))
print("Precision Score of KNeighborsClassifier is:",precision_score(y_test, prediction2,average='weighted'))
print("F1 Score is : ",f1_score(y_test, prediction2, average='weighted'))
```

```python
print(classification_report(y_test, prediction2))
print(confusion_matrix(y_test, prediction2))
```

```python
df_val = pd.DataFrame({'Actual': y_test, 'Predicted_Score': prediction2})
df_val.head(10)
```
